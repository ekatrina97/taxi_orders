#  Прогнозирование заказов такси

**Заказчик исследования** — компания «Чётенькое такси».

**Цель проекта** — прогнозировать количество заказов такси в аэропортах на следующий час, для того чтобы компаня-заказчик могла приклегать больше водителей в период пиковой нагрузки.

**Задача проекта** — построить модель машинного обучения для предсказания количества заказов на каждый следующий час в зависимости от истории заказов. Качество модели будет оцениваться с помощью метрики *RMSE*, значение которой должно быть не более 48.

**Исходные данные** лежат в файле `taxi.csv` и представлены временныи рядом: `время` — `количество заказов`.

По графику видим, что имеются значения, которые сильно выбиваются из общей картины данных. Заменим их на предыдущие во временной ряде, чтобы редкие экстремальные значения не оказывали влияния на результат предсказания модели. 

```python
# создание столбца с предыдущими во времени значениями
df['num_orders_before'] = df['num_orders'].shift(1)
```

```python
# просмотр первых строк
df.head()
```

Напишем функцию, которая заменять значениям временного ряда в зависимости от его величины. Количество заказов ограничим значением в 408.

```python
# функция ограничения максимального количества заказов
def filter_max_value(row):
    if row['num_orders'] >= 408:
        return row['num_orders_before']
    else:
        return row['num_orders']
```

```python
# применение функции
df['num_orders_filter'] = df.apply(filter_max_value, axis=1)
```

```python
# проверка фильтрации функцией
df['num_orders_filter'].max()
```

```python
# исключение колонок и переименование новой колонки
df.drop(columns=['num_orders', 'num_orders_before'], inplace=True)
df.rename(columns={'num_orders_filter': 'num_orders'}, inplace=True)
```

## Подготовка данных к обучению

Для обучения нам понадобится создать дополнительные признаки из уже имеющихся. Отдельными признаками выделим *год, месяц, число, день недели* из даты, также создадим временные ряды со смещениями (от существующего временного ряда) и будем использовать в качестве признаков.

```python
# удаление колонок, необходимых для графиков
df.drop(columns=['std', 'mean'], inplace=True)
```

```python
# функцию создания дополнительных признаков
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
# применение функции
make_features(df, 15, 9)
```

Разделим датасет на тренировочную и тестовую выборки.

```python
# деление датасета на выборки
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()

train_target = train['num_orders']
train_features = train.drop(columns=['num_orders'])

test_target = test['num_orders']
test_features = test.drop(columns=['num_orders'])
```

```python
# просмотр первых строк
train_features.head()
```

## Обучение и тестирование

Будем использовать четыре модели для обучения и сравнения результатов:
* **LGBMRegressor()**
* **CatBoostRegressor()**
* **SGDRegressor()**
* **LinearRegression()**

Сначала отберём лучшие параметры моделей с помощью *GridSearchCV()*, затем обучим модели и проверим на тестовой выборке.
Метрикой качества в нашей задаче является **RMSE**. Для модели *CatBoostClassifier()* не будем отбирать лучшие параметры, т.к. при разработке этой модели был учтён поиск оптимальных параметров.

```python
# модель LGBMRegressor()
gbm_pipe = Pipeline([('gbm', LGBMRegressor(num_leaves=30, random_state=42))])
params_gbm = [{'gbm__learning_rate': [0.01, 0.1, 1],
               'gbm__n_estimators': [20, 30]}]

tscv = TimeSeriesSplit(n_splits=8) # cv для временных рядов

gs_gbm = GridSearchCV(gbm_pipe,
                      param_grid=params_gbm,
                      scoring='neg_root_mean_squared_error',
                      cv=tscv)

gs_gbm.fit(train_features, train_target)
gbm_best_params = gs_gbm.best_params_
gbm_best_score = round(gs_gbm.best_score_, 2)
gbm_time_refit = round(gs_gbm.refit_time_, 2)

start_time = time.time()
gs_gbm.predict(train_features)
predict_time = time.time()

gbm_dict = {
    'model' : 'LGBMRegressor()',
    'best_params' : gbm_best_params,
    'RMSE': -gbm_best_score,
    'time_fit, sec' : gbm_time_refit,
    'time_predict, sec' : round(predict_time - start_time, 5)
}
gbm_dict
```

```python
# модель CatBoostRegressor()
ctb_model = ctb.CatBoostRegressor(random_state=42)

tscv = TimeSeriesSplit(n_splits=8) # cv для временных рядов

best_neg_rmse = max(cross_val_score(ctb_model, train_features,
                            train_target, cv=tscv,
                            scoring='neg_root_mean_squared_error'))
start_time = time.time()
ctb_model.fit(train_features, train_target)
fit_time = time.time()
ctb_model.predict(train_features)
predict_time = time.time()

ctb_dict = {
    'model' : 'CatBoostRegressor()',
    'best_params' : '-',
    'RMSE' : -round(best_neg_rmse, 3),
    'time_fit, sec' : round(fit_time - start_time, 2),
    'time_predict, sec' : round(predict_time - fit_time, 5)
}
ctb_dict
```

```python
# модель SGDRegressor()
sgd_pipe = Pipeline([('sgd', SGDRegressor(random_state=42))])
params_sgd = [{'sgd__alpha': [0.5, 0.1, 0.05, 0.001, 0.0001],
               'sgd__n_iter_no_change': [5, 6, 7, 8]}]

tscv = TimeSeriesSplit(n_splits=8) # cv для временных рядов

gs_sgd = GridSearchCV(sgd_pipe,
                      param_grid=params_sgd,
                      scoring='neg_root_mean_squared_error',
                      cv=tscv)

gs_sgd.fit(train_features, train_target)

sgd_best_params = gs_sgd.best_params_
sgd_best_score = round(gs_sgd.best_score_, 3)
sgd_time_refit = round(gs_sgd.refit_time_, 2)

start_time = time.time()
gs_sgd.predict(train_features)
predict_time = time.time()

sgd_dict = {
    'model' : 'SGDRegressor()',
    'best_params' : sgd_best_params,
    'RMSE': -sgd_best_score,
    'time_fit, sec' : sgd_time_refit,
    'time_predict, sec' : round(predict_time - start_time, 5)
}
sgd_dict
```

```python
# модель LinearRegression()
lnr = LinearRegression()

tscv = TimeSeriesSplit(n_splits=8) # cv для временных рядов

best_neg_rmse = max(cross_val_score(lnr, train_features,
                            train_target, cv=tscv,
                            scoring='neg_root_mean_squared_error'))
start_time = time.time()
lnr.fit(train_features, train_target)
fit_time = time.time()
lnr.predict(train_features)
predict_time = time.time()

lnr_dict = {
    'model' : 'LinearRegression()',
    'best_params' : '-',
    'RMSE' : -round(best_neg_rmse, 3),
    'time_fit, sec' : round(fit_time - start_time, 3),
    'time_predict, sec' : round(predict_time - fit_time, 5)
}
lnr_dict
```

```python
# настройка отображения формата данных
pd.set_option("display.float_format", "{:.3f}".format)
```

```python
# таблица сравнения моделей
compare_table = pd.DataFrame(data = [gbm_dict, ctb_dict, sgd_dict, lnr_dict], 
                             columns = ['model', 'RMSE', 'time_fit, sec', 'time_predict, sec'],
                             index=[1, 2, 3, 4])
compare_table
```

Лучший результат показала модель **CatBoostRegressor()**, у которой ошибка на кросс-валидации принимает самое низкое значение. Метрика качества *RMSE* составляет примерно 21. Обучается данная модель дольше всего, однако время предсказания составляет доли секунд.

Сравнение моделей произведено на основании кросс-валидации.
